# Image reading and Showing

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('../data/image_1.jpg')
img1 = cv.imread('../data/image_1.jpg',0)   # gray scale
img2 = cv.imread('../data/image_1.jpg',-1)  # with alpha level
```

## Showing image with openCV

```python
def disp(wname,im):
    """
    Do not exit using red x button of window, press any key to exit.
    """
    cv.imshow(wname,im)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
disp('image',img)
```

```python
disp('image-1',img1)
```

```python
disp('image-2',img2)
```

```python
cv.imwrite('../data/image_1_write.jpg',img)
```

## Showing  image with matplotlib

The cv2 convensons for image is \[R, G, B\] whereas in matplotlib it is \[R,G,B\]

```python
def bgr2rgb(im):
    b,g,r = cv.split(im)
    out = cv.merge([r,g,b])
    return out
```

```python
plt.figure("Image")
plt.imshow(bgr2rgb(img))
plt.show()
```

```python
plt.figure(figsize=(15,45))
plt.subplot(1,3,1)
plt.imshow(bgr2rgb(img))

plt.subplot(1,3,2)
plt.imshow(img1,cmap='gray')

plt.subplot(1,3,3)
plt.imshow(bgr2rgb(img2))

plt.tight_layout()
plt.show()
```
